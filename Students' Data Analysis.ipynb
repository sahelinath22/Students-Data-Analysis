```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student Scores.csv")
print(df.head())
```

## Data Inspection

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Dropping Unnamed Column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

## Changing Weekly Study Hours Column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","05-10")
df.head()
```

## Exploratory Data Analysis

## 1. Gender Distribution

```python
plt.figure(figsize = (6,4))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

## From the above chart we have analysed that the number of females in the data is more than the numbdr of male

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

## 2. Heatmap of Student Marks based on Parent Education

```python
plt.figure(figsize = (6,4))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents' Education and Students' score")
plt.show()
```

## From the above chart we have concluded that the education of the parents have a good impact on students' scores.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

## 3. Heatmap of Student Marks based on Parent Marital Status

```python
plt.figure(figsize = (6,4))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parents' Marital Status and Students' score")
plt.show()
```

## From the above chart we have concluded that there is no/negligible impact on the students' score due to their parents' marital status.

## 4. Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"] ]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()

```

```python
plt.figure(figsize = (10,3))
ax = sns.countplot(data = df, x = "EthnicGroup")
ax.bar_label(ax.containers[0])
plt.show()
```

## 5. Distribution of Sport Practice with Lunch Type

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(data = df, x = "PracticeSport", hue = "LunchType")
ax.bar_label(ax.containers[0])
plt.title("Distribution of Sport Practice with Lunch Type")
plt.show()
```
